```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update(
    {'font.size': 15, 'lines.markersize': 0, 'lines.linewidth': 1.2, 'axes.grid': True, 'lines.marker': '.'})
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 8), sharex=True)

Y1 = np.arange(2005, 2022)
Y2 = np.arange(2005, 2022)
Num1 = np.linspace(0, 100, len(Y1))
Num2 = np.linspace(200, 400, len(Y1))

ax1.bar(Y2, Num2, label='second')
ax1.bar(Y1, Num1, label='first')
ax1.legend()
ax2.bar(Y2, Num2)
ax2.bar(Y1, Num1)
ax1.set_ylim(min(Num2), max(Num2)*1.05)
ax2.set_ylim(min(Num1), max(Num1)*1.05)

ax2.set_xlabel('Year')
ax2.set_ylabel('Number of thousands', loc='top')
ax1.set_title('Growth of protein databases')
ax2.locator_params(nbins=len(Y2), axis='x')
# ax2.tick_params(axis='x', rotation=45)
# labels =
# ax2.set_xlabel(pdbY)

ax1.spines.bottom.set_visible(False)
ax2.spines.top.set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()
# Now, let's turn towards the cut-out slanted lines.
# We create line objects in axes coordinates, in which (0,0), (0,1),
# (1,0), and (1,1) are the four corners of the axes.
# The slanted lines themselves are markers at those locations, such that the
# lines keep their angle and position, independent of the axes size or scale
# Finally, we need to disable clipping.

d = .5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)
```
